**Assignment 7.1**

We want to determine the refractive index $n$ of a piece of glass. To this end, we measure the refraction
of light at the air-glass interface for a number of different angles of incident. The results are: 

$\phi_i$ = [10.1, 14.7, 21.5, 27.2, 31.5, 39.1, 44.6, 49.0, 54.9, 60.1, 65.5, 71.7, 76.3, 83.4, 86.5] 

$\phi_r$ = [7.6, 9.5, 16.5, 19.7, 16.4, 28.0, 27.9, 24.9, 32.3, 33.2, 31.7, 32.3, 33.6, 35.2, 35.7] 

, where $\phi_i$ is the angle of incidence and $\phi_r$ the angle of refraction. All angles are specified in degrees!
The error in the measurements are negligable for $\phi_i$ and 3 degrees for $\phi_r$.
The relation between the incident angle and the angle of refraction is given by Snellâ€™s law, which in this
case reduces to (we take the refractive index of air to be equal to 1):

$$\sin{(\phi_r)} = \frac{1}{n} \sin{(\phi_i)}$$

In the figure below the data (dots) and a theoretical relation (dotted line) is presented.

Write code that reproduces the figure **exactly**. 

```python
# =============== YOUR CODE GOES HERE =================
import numpy as np
import matplotlib.pyplot as plt

i = np.array ([10.1, 14.7, 21.5, 27.2, 31.5, 39.1, 44.6, 49.0, 54.9, 60.1, 65.5, 71.7, 76.3, 83.4, 86.5])
r = np.array ([7.6, 9.5, 16.5, 19.7, 16.4, 28.0, 27.9, 24.9, 32.3, 33.2, 31.7, 32.3, 33.6, 35.2, 35.7])
n = 1.654
swag = np.linspace(0,1,1000)
n_berekening=swag/n
i_sin = np.sin(np.radians(i))
r_sin = np.sin(np.radians(r))
fig, ax = plt.subplots(figsize=(7,3.5))
ax.errorbar(i_sin, r_sin, np.sin(np.radians(3)), fmt='o', linewidth=1.5, color='black', label = "data")
ax.plot(swag, n_berekening, ':', color='black', label = "model n=1.654")
ax.set_xlabel('sin(i)')
ax.set_ylabel('sin(r)')
ax.legend()
plt.grid()
plt.ylim(bottom=0, top=0.7)  
plt.xlim(left=0, right=1.1)
plt.show()
```

**Assignment 7.2**

Shot put (kogelstoten) is a sport where the aim is to throw a ball ($m=7.26$ kg) as far as possible. An athlete throws the ball with an intial speed $v$ under an angle $\theta=37.0$ degrees from an intial height $h_0=2.10$ m. In this assignment you make a figure that contains two plots that show both height and velocity of the trajectory of the ball for three different intial speeds: $v=10$, $v=12$, and $v=14$ m/s.

Make a script that creates a figure that consists of two subplots (see: https://matplotlib.org/stable/gallery/subplots_axes_and_figures/subplot.html#sphx-glr-gallery-subplots-axes-and-figures-subplot-py). The top plot should show the height of the ball as function of horizontal displacement. The bottom plot should show the velocity as function of the horizontal diplacement. Consider all rules that make a good graph (see reader).

```python
# =============== YOUR CODE GOES HERE =================
import numpy as np
import matplotlib.pyplot as plt

mass = 7.26
angle = 37.0
initial_height = 2.10
f_gravity = 9.81
initial_velocity = np.array([10,12,14])
time = np.linspace(0,2,100)
fig, (ax1, ax2) = plt.subplots(2, 1)


for i_initial_velocity in initial_velocity:
    vx_ball = np.cos(np.radians(angle))*i_initial_velocity
    vy_ball = np.sin(np.radians(angle))*i_initial_velocity
    
    ax_ball = 0
    ay_ball = -f_gravity

    rx_ball = 0
    ry_ball = initial_height

    v_ball = np.sqrt ((vy_ball+ay_ball*time)**2 + vx_ball**2)  
    y_ball = 0.5*ay_ball*np.power(time, 2) + vy_ball*time + ry_ball #positie = positie + vt + 1/2at^2
    x_ball = 0.5*ax_ball*np.power(time, 2) + vx_ball*time + rx_ball

    ax1.plot(x_ball, y_ball, label=f"Velocity = {i_initial_velocity}")
    ax2.plot(x_ball, v_ball, label=f"Velocity = {i_initial_velocity}")

# Grafiekje 1
ax1.set_ylabel('Height')
ax1.set_xlabel('Distance')
ax1.grid()
ax1.legend()

# Grafiekje 2
ax2.set_xlabel('Distance')
ax2.set_ylabel('Velocity')
ax2.grid()
ax2.legend()


plt.subplots_adjust(hspace = 0.5)
fig.suptitle('Focking coole natuurkunde')
ax1.set_ylim(0)
plt.show()

```
